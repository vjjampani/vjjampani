```python
import pandas as pd
```

```python
#reading the data from the url. 
url=pd.read_html('http://www.factiva.com/CP_Developer/ProductHelp/FDK/FDK20/registration/shared_code_tables/state_province_table/united_states_state_province_table.htm')
```

```python
#Loading the data into the state_df from the url
state_df=url[0]
```

```python
#printing the state_df, the below output contains : 1. Serial number 2.State code. 3. State complete name.
state_df
```

```python
state_dataframe = pd.DataFrame(state_df)
```

```python
# printingg data frame : state_dataframe
state_dataframe
```

```python
# Replace Header : 1. State/Province to US_STATE_NAME 2. code to US_STATE_CODE

state_dataframe.columns = ['US_STATE_CODE','US_STATE_NAME']
```

```python
#printing modified data frame:

state_dataframe
```

```python
# Find Any outliners in US_STATE_CODE and US_STATE_NAME
state_dataframe.describe()[['US_STATE_CODE', 'US_STATE_NAME']]
```

```python
# Fromthe above output, it seems there are no outliners in the US_STATE_CODE and US_STATE_NAME
```

```python
# Find any duplicate rows across US_STATE_CODE,US_STATE_NAMES
dupRows = state_dataframe[state_dataframe.duplicated(['US_STATE_CODE','US_STATE_NAME'])]
```

```python
#print dupRows
dupRows
```

```python
# The above output shows there are no duplicates in the "state_dataframe" 
```
